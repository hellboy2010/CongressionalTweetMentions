```python
import pandas as pd
#import xml.etree.ElementTree as ET
#import csv
import re, os, nltk
from nltk.stem.snowball import SnowballStemmer
import numpy as np
from lxml import etree
import json
from pprint import pprint

```

```python
#first way
import json
connection_file = open('data2.json', encoding="utf8")
conn_string = json.load(connection_file)
connection_file.close()



#conn_string
```

```python
connection_file2 = open('data.json', encoding="utf8")
conn_string2 = json.load(connection_file2)
connection_file2.close()
```

```python
import pandas as pd
df=pd.DataFrame.from_dict(conn_string)
df2=pd.DataFrame.from_dict(conn_string2)

print("first df:", df.size, "second df:",df2.size)
```

```python
df.head(5)
```

```python
mentions=[]
import re
for i in range(len(df['text'])):
    
    result = re.findall("@(\w{1,15})", df['text'][i])
    mentions.append(result)
print(mentions)
```

```python
mentions2=[]
import re
for i in range(len(df2['text'])):
    
    result = re.findall("@(\w{1,15})", df2['text'][i])
    mentions2.append(result)
```

```python
mentions_to_pd = pd.DataFrame(mentions)
mentions2_to_pd = pd.DataFrame(mentions2)
```

```python
result_day1 = pd.concat([df, mentions_to_pd], axis=1)
result_day2 = pd.concat([df2, mentions2_to_pd], axis=1)
```

```python
final=result_day1.append(result_day2)
```

```python
final.head(10)
final.to_csv('tweets.csv')
```
